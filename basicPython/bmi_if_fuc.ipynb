```python
# 주어진 범위내의 짝수만 반환하는 함수 if문 예시
def get_even(first,second):
    list =[]
    for i in range(first,second+1):
        if i%2 ==0:
            list.append(i)
    return list
result = get_even(1,10)
print(result)
```

```python
# 숫자 맞추기 게임 함수 while문 사용
def quiz(_input):
    count =0 
    number= None
    while _input !=number:
        number = int(input('값을 입력하시오')) 
        if number > _input:
            print('입력하시 값이 더 큼')
            count+=1
        elif number < _input:
            print('입력하시 값이 더 작음')
            count+=1
    print(f'총{count}횟수만큼 시도 후 정답을 얻음')  
quiz(50)     
```

```python
# 정규화 =(값 - 최소값)/(최대값 -최대값) 해당 범위의 값으로 균일하게 만드는 것
def normalize_data(list):
    result =[]
    _min = min(list)
    _max= max(list)
    for i in list:
        result.append((i-_min)/(_max-_min))
    return result

_result = normalize_data([10,20,30,40,50])
print(_result)
```
