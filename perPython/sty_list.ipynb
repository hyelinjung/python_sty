```python
# 파일을 읽어옴->파일이 없을 경우 에러잡고 해당 txt파일 인코딩
try:
    file = open('./test.txt','r',encoding='UTF-8')
    content = file.read()
    print(content)
except FileNotFoundError as e:
    print(f'오류내용은:{e}입니다.')
finally:
    file.close()
    print('파일을 닫음')
```

```python
#리스트 내 최대값과 최소값 찾기
nums =[1, 2, 3, 4, 5, 6, 7]
max_1=nums[0]
min_2=nums[0]
for i in range(0,len(nums)):

  if nums[i]> max_1:
    max_1= nums[i]
  elif nums[i]< min_2:
        min_2 = nums[i]

print(min_2)
print(max_1)
```

```python
#리스트 요소 합계 구하기
nums = [1,2,3,4,5]

a=0
for i in nums:
    a+=i
    
print(f'모든 요소의 합계는 {a}')
```

```python
#주어진 리스트의 요소를 역순으로 출력하는
nums = [1,2,3,4,5]
len = len(nums)-1
a =[]
while(len>=0):
  a.append(nums[len])
  print(f'a의 값으로 {a}')
  len-=1

print(f'a의 인자값 {a}')
```
