```python
#입력값이 운년인지 판단하는 함수->융년 4로 나누어 떨어지고나 400으로 나누어 떨어짐
# 100으로 나누어 떨어지면 평년
def month():
    _input = int(input('년도를 입력하시오'))
    if len(str(_input))<3:
        print('입력값은 4자리 수')
    elif _input%4==0 or _input%400 ==0:
        print('해당 년도는 운년입니다')
    elif _input%100==0:
        print('해당 년도는 평년입니다.')
    else:
        print('어느 곳에도 속하지 않음')
        
    
month()
```

```python
# 1부터 100까지의 숫자 중에서 8의 배수이지만 12의 배수는 아닌 수를 모두 출력하는 함수

def find_8_12():
    for i in range(1,101):
        if i%8==0:
            if i%12==0:
                continue
            else:
                print(i)
            
find_8_12()
```

```python
# ['dot','perls']라는 리스트가 있을 때. 리스트의 맨 앞에 'python'을 삽입하고 'perls'를 제거하여 ['python','dot']를 만드시오

def list_add(origin):
    origin.insert(0,'python')
    origin.remove('perls')
    return origin

list_add(['dot','perls'])
```

```python
# 리스트를 오름차순으로 정렬한 후 출력하고. 다시 내림차순으로 정렬하여 출력하는 프로그램 [200,300,100,50]

list =[200,100,50]
print(list.sort())
```

```python
# 다음 리스트를 문자열 길이에 따라 오름차순으로 정렬하세요 ['apple','banana','cherry','date','elderberry']

list = ['apple','banana','cherry','date','elderberry']
result = sorted(list,key=lambda x: len(x))
print(result)
```
