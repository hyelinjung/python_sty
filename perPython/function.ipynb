```python
# 매개값으로 숫자 리스트를 받아 평균을 반환하는 함수를 코딩
def getAverage(numbers):
    sum=0
    for i in numbers:
        sum +=i
        print(f'sum의 값:{sum}')
        
    return int(sum/len(numbers))


print(f'모든요소의 평균값은: {getAverage([5,6,19,10])}')
```

```python
# 주어진 데이터 중에 짝수만 반환하는 함수를 코팅하시오 -> 매개값은 리스트
import random
def filter_even(list):
    newList=[]
    # print(f'리스트1:{list}')
    for i in list:
        if i%2 ==0:
            # print('짝수')
            newList.append(i)
            # print(newList)
    return newList     
# print(filter_even([4,6,7,8,9]))

def random_list(nm):
    setList=[]
    while(nm>0):
        a=random.randint(1,30)
        setList.append(a)
        nm-=1
    print(f'만들어진 리스트:{setList}')
    return filter_even(setList)  

a=random_list(5)   
print(a)
```
