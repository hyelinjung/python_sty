```python
#주어진 튜플에서 특정 요소의 개수를 세는 프로그램->2의 개수수
numbers = (1, 2, 3, 4, 2, 5, 2, 6, 7, 2)
count=0
for i in numbers:
    if i==2:
       count+=1

print(f'2의 총개수는 {count}') 

```

```python
# 딕션너리에 요소 추가와 평균 점수르 계산
students = {"Alice": 85, "Bob": 92, "Charlie": 78, "David": 95}
students.__setitem__('lin',30,)
students_sum=0
everage=0
for i in students.values():
    students_sum +=i



everage = int(students_sum/len(students))
print(f'학생들의 평균값은 {everage}')
```

```python
# 딕셔너리를 사용해 주문 번호를 입력 시 값을 요청하는 프로그램
menu = {"burger": 5000, "pizza": 12000, "salad": 8000, "soda": 1500} 
menu_key_list= list(menu.keys())
customer_input=int(input('0~3 입력하시오'))

# 들어오는 입력값은 문자열
# print(f'들어오는 입력값의 타입{type(customer_input)}')
price= 0

if customer_input==0:
    price=menu.get(menu_key_list[0])
elif customer_input==1:
    price=menu.get(menu_key_list[1])
elif customer_input==2:
    price=menu.get(menu_key_list[2])
else:
    price=menu.get(menu_key_list[3])   

print(f'선택하신 음식[{menu_key_list[customer_input]}]의 값은 {price}입니다') 

# 'dict_keys' object is not subscriptable -> 딕셔너리에서의 키를 리스트가 아니라 뷰 객체로 반환해 인덱싱하려해 오류 발생
```
