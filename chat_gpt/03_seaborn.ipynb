```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 선그래프 lineplot() -시간에 따른 변화난 두 변수 간의 관계를 나타내는 선 그래프를 그림

# 예시 데이터
data = sns.load_dataset('tips')
# 선 그래프 그리기
sns.lineplot(x='day',y='total_bill',data=data)
# 그래프 제목과 레이블 추가
plt.title('선그래프')
plt.xlabel('day of the week')
plt.ylabel('totla bill')

plt.show()
```

```python
# 막대 그래프 ->카테고리별 평균값을 시각화하는 데 사용
sns.barplot(x='day',y='total_bill',data=data)
plt.show()
```

```python
# heatmap -> 상관 관계나 2D 데이터를 시각화하는 데 유용함

# 피벗 테이블을 사용해 데이터를 변환
data = sns.load_dataset('flights')
pivot_data = data.pivot_table(index='month',columns='year',values='passengers')
sns.heatmap(pivot_data, annot=True,cmap='YlGnBu',fmt='d')
plt.show()
```
