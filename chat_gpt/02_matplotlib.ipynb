```python
# 데이터 시각화를 위한 라이브러리
import matplotlib.pyplot as plt 
```

```python
# plot() 선그래프
x=[1,2,3,4,5]
y=[2,4,6,8,10]

# 선 그래프 그리기
plt.plot(x,y)
plt.title('Line Graph Example')
plt.xlabel('X-label')
plt.ylabel('Y-label')
plt.show()
```

```python
# 산점도 -데이터를 점으로 표현
plt.scatter(x,y,color='red')
plt.title('Scatter Plot Example')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
# bar() 막대 그래프 . 각 카테고리에 대해 막대를 그려 데이터를 비교할 수있다.

# 데이터 준비
categories =['A','B','C','D']
values =[3,7,2,5]

# 막대 그래프 그리기
plt.bar(categories,values)
plt.show()
```

```python
# hist() '데이터의 분포를 시각화하는데 사용'되는 히스토그램을 그릴때
import numpy as np

# 데이터 준비-랜덤으로 1000개 데이터 생성
data = np.random.randn(1000)

# 히스토그램 그리기
plt.hist(data, bins=30, edgecolor='black')
plt.show()
```
