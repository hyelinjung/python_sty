```python
# 클래스 정의
class Person:
    def __init__(self,name,age):
        self.name=name
        self.age =age
    # 메서드 정의
    def greet(self):
        print(f'hello, my name is {self.name} and i am {self.age} years old')
        
# 객체생성 및 호출
p1 = Person('홍길동',20)
p2 = Person('김철수',30)

p1.greet()
p2.greet()
```

```python
# 클래스 변수와 인스턴스 변수
# 클래스 변수는 클래스에 속하며 모든 객체가 공유하고 .iv는 객체마다 도유한 값을 가진다.

class Car:
    # 클래스 변수
    wheels=4
    def __init__(self,make,model):
        self.make=make
        self.model=model
    def display_info(self):
        print(f'this is a {self.make}{self.model} with {Car.wheels}')
        
c1 = Car('toyota','corolla')
c1.display_info()

# 클래스 변수 수정
Car.wheels = 3
c1.display_info()
```

```python
# 상속

# 부모 클래스
class Animal:
    def __init__(self,name):
        self.name = name
    def speak(self):
        print(f'{self.name} makes a sound')
        
# 자식 클래스
class Dog(Animal):
    def __init__(self, name,breed):
        super().__init__(name)
        self.breed = breed
        
    def speak(self):
        print(f'{self.name} barks')
        
# 객체 생성
dog = Dog('buddy','golden re')
dog.speak()
```

```python
# 메소드 오버로딩 유사 구현
class Calulator:
    def add(self,a,b=None):
        if b is None:
            return a+a
        else:
            return a+b

c1 = Calulator()
c1.add(5)
# c1.add(5,9)
```

```python
# 학생 클래스 만들기
# 학생의 이름,나이,성적을 속성으로 가지고, 학생의 성적에 따라 등급을 반환하는 클래스를 만드시오
# 90이상 A 80이상 B, 70이상 C,60이상 D, 그외 F
class Student:
    def __init__(self,name,age,grade):
        self.name = name
        self.age = age
        self.grade = grade
        
    def get_grade(self):
        if self.grade >90:
            return 'A'
        elif self.grade>80:
            return 'B'
        elif self.grade>70:
            return 'C'
        elif self.grade >60:
            return 'D'
        else:
            return 'F'
        
s1 = Student('홍길동',20,75)
print(s1.get_grade())
```

```python
# 책을(저자,제목,대출 여부를 속성으로 가지고 있음) 대출과 반납의 기능을 가지고 있는
# 도서관 클래스를 만드시오
class Book:
    def __init__(self,title,author,flag=0):
        self.title = title
        self.author =author
        self.flag = flag
    
    # 대여가능 여부
    def is_available(self):
        if self.flag ==0:
            return True
        else:
            return False
        
class Library:
    
    def __init__(self):
        self.list =[]
    # 도서 추가
    def add_book(self,book):
        self.list.append(book)
    # 도서 대여
    def get_book(self,book):
        if book.is_available():
            print('대출가능')
            book.flag=1
        else:
            print('대출중...')
            
    # 도서 반납
    def refun(self,book):
        book.flag=0
        print('반납되었습니다.')
        
b1 =Book('1번책','홍길동')
b2 =Book('2번책','김철수')
b3 =Book('3번책','레이먼')
lib1 = Library()
lib1.add_book(b1)
lib1.get_book(b1)
```
